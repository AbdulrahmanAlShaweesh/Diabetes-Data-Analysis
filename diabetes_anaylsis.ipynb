```python
x = df['Glucose']
y =  list(range(len(x)))

plt.style.use('ggplot')
plt.hist(x, bins=50, density=True, color='blue', alpha=0.5)
plt.xlabel('Glucose level') 
plt.ylabel('number of pation')
plt.title('Glucose Vs. number of pations')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
y = df['Outcome'].value_counts()
x = [0, 1]

plt.bar(x, y, width=0.2, color='k', alpha=.5)
plt.xlabel('resut')
plt.ylabel('number number of pations and non')
plt.title('resut Vs. number of people pations and none')
plt.show()
```

```python
plt.pie(df['Outcome'].value_counts(), labels=['None Diabetes', 'Diabetes'], autopct='%1.1f%%')
plt.title('Diabetes Vs. None Diabetes')
plt.show()
```

```python
sns.countplot(data=df, x=df['Pregnancies'], hue='Outcome')
plt.xlabel('')
plt.legend(title='Resute')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Diabetes and none Diabetes')
plt.title('number of Pregnancies Vs. Diabetes and none Diabetes')
plt.show()
```
